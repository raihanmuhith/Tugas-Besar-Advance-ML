```python
# buat dataframe untuk looping

df_x = df.drop(['Date','Location'],1)
```

```python
arr = df_x.axes[1].tolist()
```

```python
# membagi variabel kategorikal dan numerik

list_kategorik = ['RainToday','RainTomorrow','WindGustDir','WindDir9am','WindDir3pm']
list_numerik = [x for x in arr if x not in list_kategorik]
```

```python
#mereplace semua missing value variabel numerik dengan nilai rata-rata variabel tsb.

for i in list_numerik:
    df[i].fillna(round(df[i].mean()),inplace=True)
```

```python
#mereplace semua missing value variabel kategorikal dengan nilai modus variabel tsb.

for j in list_kategorik:
    df[j].fillna(df[j].mode()[0],inplace=True)
```

```python
#fungsi mencari nilai lower bound dan upper bound dari suatu dataframe

def get_lower_upper_bound(x):
    q1 = np.percentile(x,25)
    q3 = np.percentile(x,75)
    iqr = q3 - q1
    lb = q1 - (iqr*1.5)
    ub = q3 + (iqr*1.5)
    return lb,ub
```

```python
#fungsi mencari outliers dari suatu dataframe

def get_outliers_iqr(x):
    lb,ub = get_lower_upper_bound(x)
    return x[np.where((x>ub) | (x<lb))]
```

```python
for i in df_x:
    print("banyaknya bulatan hitam dari variabel",i,":",len(get_outliers_iqr(df[i].values)))
```

```python
# drop outliers with iqr

def remove_outliers_iqr(df):
    Lb,Ub = get_lower_upper_bound(df)
    df_final = df[(df>=Lb) & (df<=Ub)]
    return df_final
```

```python
# removing outliers

removed_outliers = df.copy()
for i in df_x.drop(['RainToday','RainTomorrow','Rainfall','Evaporation'],1):
    removed_outliers[i] = remove_outliers_iqr(df[i].copy())

removed_outliers
```

```python
removed_outliers.isna().sum()
```

```python
removed_outliers.dropna(inplace=True)
```

```python
# Normalize with MinMax Scalling

def normalize(x):
    return (x - x.min())/(x.max() - x.min())
```
